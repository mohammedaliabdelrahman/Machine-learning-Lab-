```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.histplot(df['Customer_Age'], bins=20, kde=True)
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')

plt.subplot(1,2,2)
sns.boxplot(y=df['Customer_Age'])
plt.title('Customer Age Boxplot')
plt.tight_layout()
plt.show()

print(f"Mean Age: {df['Customer_Age'].mean():.2f}")
print(f"Median Age: {df['Customer_Age'].median():.2f}")
print(f"Min Age: {df['Customer_Age'].min()}")
print(f"Max Age: {df['Customer_Age'].max()}")
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.histplot(df['Credit_Score'], bins=20, kde=True, color='orange')
plt.title('Credit Score Distribution')
plt.xlabel('Credit Score')
plt.ylabel('Count')

plt.subplot(1,2,2)
sns.boxplot(y=df['Credit_Score'], color='orange')
plt.title('Credit Score Boxplot')
plt.tight_layout()
plt.show()

print(f"Mean Credit Score: {df['Credit_Score'].mean():.2f}")
print(f"Median Credit Score: {df['Credit_Score'].median():.2f}")
print(f"Min Credit Score: {df['Credit_Score'].min()}")
print(f"Max Credit Score: {df['Credit_Score'].max()}")
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.histplot(df['Annual_Income'], bins=20, kde=True, color='green')
plt.title('Annual Income Distribution')
plt.xlabel('Annual Income ($)')
plt.ylabel('Count')

plt.subplot(1,2,2)
sns.boxplot(y=df['Annual_Income'], color='green')
plt.title('Annual Income Boxplot')
plt.tight_layout()
plt.show()

print(f"Mean Income: ${df['Annual_Income'].mean():.2f}")
print(f"Median Income: ${df['Annual_Income'].median():.2f}")
print(f"Min Income: ${df['Annual_Income'].min()}")
print(f"Max Income: ${df['Annual_Income'].max()}")
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.histplot(df['Purchase_Amount'], bins=20, kde=True, color='purple')
plt.title('Purchase Amount Distribution')
plt.xlabel('Purchase Amount ($)')
plt.ylabel('Count')

plt.subplot(1,2,2)
sns.boxplot(y=df['Purchase_Amount'], color='purple')
plt.title('Purchase Amount Boxplot')
plt.tight_layout()
plt.show()

print(f"Mean Purchase: ${df['Purchase_Amount'].mean():.2f}")
print(f"Median Purchase: ${df['Purchase_Amount'].median():.2f}")
print(f"Min Purchase: ${df['Purchase_Amount'].min()}")
print(f"Max Purchase: ${df['Purchase_Amount'].max()}")
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
gender_counts = df['Gender'].value_counts()
gender_counts.plot(kind='bar', color=['blue', 'pink', 'gray'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)

plt.subplot(1,2,2)
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['blue', 'pink', 'gray'])
plt.title('Gender Percentage')
plt.tight_layout()
plt.show()

print(gender_counts)
```

```python
plt.figure(figsize=(12,5))
category_counts = df['Purchase_Category'].value_counts()
category_counts.plot(kind='bar', color='skyblue')
plt.title('Purchase Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(category_counts)
```

```python
plt.figure(figsize=(10,5))
provider_counts = df['BNPL_Provider'].value_counts()
provider_counts.plot(kind='bar', color=['red', 'blue', 'green', 'orange'])
plt.title('BNPL Provider Distribution')
plt.xlabel('Provider')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print(provider_counts)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
status_counts = df['Repayment_Status'].value_counts()
colors = ['green', 'orange', 'red']
status_counts.plot(kind='bar', color=colors)
plt.title('Repayment Status Distribution')
plt.xlabel('Status')
plt.ylabel('Count')
plt.xticks(rotation=0)

plt.subplot(1,2,2)
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', colors=colors)
plt.title('Repayment Percentage')
plt.tight_layout()
plt.show()

print(status_counts)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Credit_Score', y='Purchase_Amount', data=df, alpha=0.5, hue='Repayment_Status')
plt.title('Credit Score vs Purchase Amount')
plt.xlabel('Credit Score')
plt.ylabel('Purchase Amount ($)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Customer_Age', y='Purchase_Amount', data=df, alpha=0.5, color='green')
plt.title('Age vs Purchase Amount')
plt.xlabel('Age')
plt.ylabel('Purchase Amount ($)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Annual_Income', y='Purchase_Amount', data=df, alpha=0.5, color='purple')
plt.title('Annual Income vs Purchase Amount')
plt.xlabel('Annual Income ($)')
plt.ylabel('Purchase Amount ($)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Repayment_Status', y='Credit_Score', data=df)
plt.title('Credit Score by Repayment Status')
plt.xlabel('Status')
plt.ylabel('Credit Score')
plt.xticks(rotation=0)
plt.show()

credit_by_status = df.groupby('Repayment_Status')['Credit_Score'].mean()
print(credit_by_status)
```

```python
plt.figure(figsize=(12,5))
avg_by_category = df.groupby('Purchase_Category')['Purchase_Amount'].mean().sort_values(ascending=False)
avg_by_category.plot(kind='bar', color='skyblue')
plt.title('Average Purchase Amount by Category')
plt.xlabel('Category')
plt.ylabel('Average Amount ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(avg_by_category)
```

```python
plt.figure(figsize=(8,5))
avg_by_provider = df.groupby('BNPL_Provider')['Purchase_Amount'].mean().sort_values(ascending=False)
avg_by_provider.plot(kind='bar', color='lightcoral')
plt.title('Average Purchase Amount by Provider')
plt.xlabel('Provider')
plt.ylabel('Average Amount ($)')
plt.xticks(rotation=0)
plt.show()

print(avg_by_provider)
```

```python
gender_status = pd.crosstab(df['Gender'], df['Repayment_Status'], normalize='index') * 100
gender_status.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Repayment Status by Gender')
plt.xlabel('Gender')
plt.ylabel('Percentage (%)')
plt.legend(title='Status')
plt.xticks(rotation=0)
plt.show()

print(gender_status)
```

```python
numerical_cols = ['Customer_Age', 'Annual_Income', 'Credit_Score', 'Purchase_Amount']
correlation = df[numerical_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, square=True)
plt.title('Correlation Matrix')
plt.show()

print(correlation)
```

```python
df['Age_Group'] = pd.cut(df['Customer_Age'], bins=[0,25,35,50,65,100], labels=['18-25','26-35','36-50','51-65','65+'])
age_default = df.groupby('Age_Group')['Repayment_Status'].apply(lambda x: (x=='Defaulted').mean()*100)
age_default.plot(kind='bar', figsize=(8,5), color='red')
plt.title('Default Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Default Rate (%)')
plt.xticks(rotation=0)
plt.show()

print(age_default)
```

```python
df['Income_Bracket'] = pd.cut(df['Annual_Income'], bins=5, labels=['Very Low','Low','Medium','High','Very High'])
income_default = df.groupby('Income_Bracket')['Repayment_Status'].apply(lambda x: (x=='Defaulted').mean()*100)
income_default.plot(kind='bar', figsize=(8,5), color='orange')
plt.title('Default Rate by Income Bracket')
plt.xlabel('Income Bracket')
plt.ylabel('Default Rate (%)')
plt.xticks(rotation=0)
plt.show()

print(income_default)
```

```python
total = len(df)
paid = len(df[df['Repayment_Status']=='Paid On Time'])
late = len(df[df['Repayment_Status']=='Late Payment'])
default = len(df[df['Repayment_Status']=='Defaulted'])

print("="*50)
print("KEY FINDINGS")
print("="*50)
print(f"Total Transactions: {total}")
print(f"Paid On Time: {paid} ({paid/total*100:.2f}%)")
print(f"Late Payment: {late} ({late/total*100:.2f}%)")
print(f"Defaulted: {default} ({default/total*100:.2f}%)")
print(f"\nAverage Age: {df['Customer_Age'].mean():.1f}")
print(f"Average Income: ${df['Annual_Income'].mean():.2f}")
print(f"Average Credit Score: {df['Credit_Score'].mean():.1f}")
print(f"Average Purchase: ${df['Purchase_Amount'].mean():.2f}")
print(f"\nMost Popular Category: {df['Purchase_Category'].mode()[0]}")
print(f"Most Used Provider: {df['BNPL_Provider'].mode()[0]}")
```
